```python
anthro = pd.read_csv('anthro.csv')
anthro
```

```python
sns.boxplot(data=anthro)
sns.swarmplot(data=anthro,color=".2", size=3.6)
#SE_ax.set_xticklabels(rotation=90)
plt.ylabel('Data Distribution')
plt.title('Anthropomorphism Rating')
plt.ylim(1, 5.2)
plt.show()
```

```python
my_plt = [0,0,6,16,2]
plt.bar(range(1,len(my_plt)+1), my_plt)
plt.xlabel('Rating')
plt.ylabel('Values')
plt.title('Responsiveness')
```

```python
time = [20, 4, 0]
position = ['Yes', 'Unsure', 'No']
plt.xlabel('Rating')
plt.ylabel('Values')
plt.title('Recognizability')
plt.bar(position, time, color="darkblue")
```

```python
like = pd.read_csv('like.csv')
like
```

```python
sns.boxplot(data=like)
sns.swarmplot(data=like,color=".2", size=3.6)
#SE_ax.set_xticklabels(rotation=90)
plt.ylabel('Data Distribution')
plt.title('Likeability Rating')
plt.ylim(1, 5.2)
plt.show()
```

```python
anthro_mean =anthro.mean()
anthro_mean_df =pd.DataFrame(anthro_mean)
anthro_mean_df.transpose()
```

```python
sns.barplot(data=anthro_mean_df.transpose())
plt.ylim(0,5)
plt.ylabel('Rating')
plt.title('Anthropomorphism Mean Rating')
```
