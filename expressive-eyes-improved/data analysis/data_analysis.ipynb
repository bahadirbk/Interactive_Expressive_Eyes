```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = pd.read_csv('results.csv')
results
```

```python
plt.hist(results['Social Rapport'],color = "blue")
plt.xlabel('Value')
plt.ylabel('Occurence')
plt.title('Building Social Rapport')
```

```python
anthro = pd.read_csv('anthro.csv')
anthro
```

```python
sns.boxplot(data=anthro)
sns.swarmplot(data=anthro,color=".2", size=3.6)
#SE_ax.set_xticklabels(rotation=90)
plt.ylabel('Rating')
plt.title('Anthropomorphism')
plt.show()
```

```python
like = pd.read_csv('like.csv')
like
```

```python
sns.boxplot(data=like)
sns.swarmplot(data=like,color=".2", size=3.6)
#SE_ax.set_xticklabels(rotation=90)
plt.ylabel('Rating')
plt.title('Likeability')
plt.show()
```

```python
anthro_mean =anthro.mean()
anthro_mean_df =pd.DataFrame(anthro_mean)
anthro_mean_df.transpose()
```

```python
sns.barplot(data=anthro_mean_df.transpose())
plt.ylim(0,5)
plt.ylabel('Rating')
plt.title('Anthropomorphism Mean Rating')
```

```python
like_mean =like.mean()
like_mean_df =pd.DataFrame(like_mean)
like_mean_df.transpose()
```

```python
sns.barplot(data=like_mean_df.transpose())
plt.ylim(0,5)
plt.ylabel('Rating')
plt.title('Likeability Mean Rating')
```
